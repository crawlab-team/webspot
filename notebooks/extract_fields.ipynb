```python
from webspot.detect.detectors.plain_list import PlainListDetector
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
from networkx.drawing.nx_agraph import graphviz_layout
```

```python
json_path = '/Users/marvzhang/projects/tikazyq/webspot/data/quotes.toscrape.com/json/http___quotes_toscrape_com_.json'
html_path = '/Users/marvzhang/projects/tikazyq/webspot/data/quotes.toscrape.com/html/http___quotes_toscrape_com_.html'
detector = PlainListDetector(json_path=json_path, html_path=html_path)
detector.run()
```

```python
d = detector
```

```python
l = detector.graph_loader
```

```python
g = detector.graph_loader.g_dgl
```

```python
list_node_id = d.results[0].list_node.id
list_node_id
```

```python
g.subgraph(list_node_id)
```

```python
min_node_id = min([n.id for n in l.nodes_])
g_nx = g.to_networkx()
g_nx.remove_nodes_from(range(min_node_id))
g_nx.nodes()
```

```python
g_nx.edges()
```

```python
pos = graphviz_layout(g_nx, prog='dot')
plt.figure(figsize=(20, 6))
nx.draw(g_nx, pos, with_labels=True)
```

```python
nx.is_tree(g_nx)
```

```python
list(g_nx.successors(16))
```

```python
tree_data = nx.tree_data(g_nx, 16)
```

```python
g_nx_tree = nx.tree_graph(tree_data)
```

```python
pos = graphviz_layout(g_nx_tree, prog='dot')
plt.figure(figsize=(20, 6))
nx.draw(g_nx_tree, pos, with_labels=True)
```
